```python
import dhlab.nbtext as nb
import pandas as pd
```

```python
metanob = nb.book_corpus(period = (1810, 1910), lang='nob', limit = 20000)
```

```python
metadan = nb.book_corpus(period = (1810, 1910), lang = 'dan', limit = 20000)
```

```python
metanno = nb.book_corpus(period = (1810, 1910), lang = 'nno', limit = 20000)
```

```python
metanno
```

```python
metadan
```

```python
metadan['lang'] = 'dan'
```

```python
metanob['lang'] = 'nob'
```

```python
metanno['lang'] = 'nno'
```

```python
meta = pd.concat([metanob, metadan, metanno])
```

```python
meta['url'] = "https://urn.nb.no/URN:NBN:no-nb_digibok_" + meta['urn'].astype(str)
```

```python
meta = meta.set_index('urn')
```

```python
meta.to_csv('meta_1800_nob_nno_dan.csv')
```
